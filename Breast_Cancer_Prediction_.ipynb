**Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,roc_auc_score,roc_curve
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from warnings import filterwarnings
filterwarnings('ignore')
```

**Load Dataset**

```python
data = pd.read_csv("cancer.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
# drop the unnamed and id columns.
data = data.drop(columns=['Unnamed: 32', 'id'])
```

```python
#Encoding categorical data values
labelencoder = LabelEncoder()
data['diagnosis'] = labelencoder.fit_transform(data['diagnosis'])
data['diagnosis'].value_counts()
```

```python
sns.countplot(x='diagnosis', data=data, palette='Set2')
```

**Exploratory Data Analysis (EDA)**

```python
# Data visualization to create histogram
data.hist(figsize=(11,11),edgecolor = 'black')
plt.show()
```

```python
plt.figure(figsize=(25,12))
sns.heatmap(data.corr(),annot = True, cmap = "cubehelix_r")
plt.show()
```

**Split the Data**

```python
X = data.drop('diagnosis', axis=1)
Y = data['diagnosis']
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

**Model selection**

```python
#1.SVM
svm = SVC()
svm.fit(x_train,y_train)
prediction1 = svm.predict(x_test)
print("The Accuracy of SVM is: ",metrics.accuracy_score(prediction1, y_test))
```

```python
#3.Decision Tree 
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
prediction3 = dt.predict(x_test)
print("The Accuracy of DT is: ",metrics.accuracy_score(prediction3, y_test))
```

```python
#3.KNN
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
prediction3 = knn.predict(x_test)
print("The Accuracy of KNN is: ",metrics.accuracy_score(prediction3, y_test))
```

```python
#4.Logistic Regression
lr = LogisticRegression()
lr.fit(x_train,y_train)
prediction4 = lr.predict(x_test)
print("The Accuracy of KNN is: ",metrics.accuracy_score(prediction4, y_test))
```

**=> Logistic Regression has the highest accuracy score 0.97076**

```python
#Classification report of Logistic Regression
print(classification_report(prediction4, y_test))
```

```python
#Confusion matrix
print(confusion_matrix(prediction4, y_test))
```

```python
#create ROC curve
fpr, tpr, _ = metrics.roc_curve(prediction4, y_test)
auc = metrics.roc_auc_score(prediction4, y_test)

plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.show()
```

**Using boosting ensemble model to imporve the accuracy score**

```python
RFC = AdaBoostClassifier(base_estimator=LogisticRegression())
RFC.fit(x_train,y_train)
predict = RFC.predict(x_test)
print(classification_report(y_test,predict))
```

Now the accuracy of the module increased to 0.98
